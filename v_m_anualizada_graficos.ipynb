### Intervalo de 2010 a 2013 no Brasil com dados da ANTAQ

```python
fig, ax = plt.subplots()
plt.scatter(data_0['Ano'], data_0['Total de Movimentação Contêineres (Líquido)\nem toneladas (t)'])
#plt.xticks(data_0['Ano'])
#plt.yticks(data_0['Total de Movimentação Contêineres (Líquido)\nem toneladas (t)'])
plt.plot(data_0['Ano'], data_0['Total de Movimentação Contêineres (Líquido)\nem toneladas (t)'])
plt.show()
```
